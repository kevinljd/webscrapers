```python
from bs4 import BeautifulSoup
import requests
import pandas
import time

base_url = 'http://example.webscraping.com/places/default/index/'

# r = requests.get(first_pg)
# c = r.content
# soup = BeautifulSoup(c)
# # print(soup.prettify())

index = 0
l = []
while True:
    r = requests.get(base_url + str(index))
    c = r.content
    soup = BeautifulSoup(c)
    
    
    all = soup.find_all('tr')
    if len(all) == 0:
        break
    
#     if index == 2:
#         break
        
    divs = []
    for row in all:
        divs.append(row.find_all('div'))

    # print(divs[0][0].text)
    # print(divs)
    for row in divs:
        for div in row:
            print(div.text)
            d = {}
            d['Page'] = index + 1
            d['Countries'] = div.text
            l.append(d)
            
    index = index + 1
    print('\n Page:' + str(index) + '\n')
    time.sleep(5)

# print(l)
```

```python
df = pandas.DataFrame(l)
print(df)
```

```python
df.to_csv("Countries.csv")
```
