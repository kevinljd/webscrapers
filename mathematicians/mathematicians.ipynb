```python
import requests
from bs4 import BeautifulSoup

r = requests.get('http://www.fabpedigree.com/james/mathmen.htm')
c = r.content
soup = BeautifulSoup(c, 'html.parser')

all = soup.find_all('li')


s = set()
for mathematician in all:
    for name in mathematician.text.split('\n'):
        if name != '':
            s.add(name)

for name in s:
    print(name)
```

```python
import pandas

l = []
for name in s:
    d = {}
    d['Names'] = name
    l.append(d)

df = pandas.DataFrame(l)
df
```
